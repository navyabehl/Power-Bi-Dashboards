```python
menu = {
    "Burger": 100,
    "Pizza": 250,
    "Pasta": 150,
    "Fries": 80,
    "Coke": 40
}

order={}

menu_list=list(menu.items())

print("WELCOME TO GOLDEN FORK FOOD DELIVERYðŸ’›")
print("HERE'S OUR MENU:\n")
for i,(item, price) in enumerate(menu_list, start=1):
    print(f"{i}.{item}= â‚¹{price}")


while True:
    choice=input("\nEnter the item number you want to order (or type 'done' to finish)").title()
    if choice.lower() == "done":
        break
    if choice.isdigit():
        item_index = int(choice)-1
        if 0 <= item_index < len(menu_list):
            item_name, item_price = menu_list[item_index]
            qty=int(input(f"How many {item_name}s would you like"))
            if item_name in order:
                order[item_name] += qty
            else:
                order[item_name] = qty
        else:
            print("Invalid item number")
    else:
        print("Please enter a valid number")

Delivery_option=input("Would you like Delivery or Pickup? Choose D or P:").strip().upper()
if Delivery_option == "D":
    address = input("Please enter your delivery address: ")
    delivery_charge = 40
else:
    address = "Store Pickup"
    delivery_charge = 0


print("\n ðŸ§¾ Order Summary:\n")
total = 0
for item, qty in order.items():
    price = menu[item] * qty
    total += price
    print(f"{item} x {qty} = â‚¹{price}")

print(f"\nDelivery Charge: â‚¹{delivery_charge}")
total += delivery_charge
print(f"\nTotal Amount: â‚¹{total}")
print(f"Delivery Option: {'Delivery' if Delivery_option == 'D' else 'Pickup'}")
print(f"Address: {address}")


confirm = input("\nDo you want to confirm the order? Choose Y or N: ").strip().upper()
if confirm == "Y":
    print("\nðŸŽ‰ Thank you! Your order has been placed.")
else:
    print("\nâŒ Order cancelled.")

```

```python
import random
import string

# Full Menu with Categories
menu = {
    "Meals": {
        "Burger": 100,
        "Pizza": 250,
        "Pasta": 150
    },
    "Snacks": {
        "Fries": 80,
        "Nachos": 120,
        "Sandwich": 90
    },
    "Drinks": {
        "Coke": 40,
        "Lemonade": 50,
        "Iced Tea": 60
    }
}

def generate_order_id():
    return ''.join(random.choices(string.ascii_uppercase + string.digits, k=6))

while True:
    order = {}

    print("\nWELCOME TO GOLDEN FORK FOOD DELIVERYðŸ’›")
    name = input("Please enter your name: ").strip()
    order_id = generate_order_id()

    print("\nHERE'S OUR MENU:\n")
    menu_list = []
    count = 1
    for category, items in menu.items():
        print(f"\nðŸ“‚ {category}:")
        for item, price in items.items():
            menu_list.append((item, price))
            print(f"{count}. {item} = â‚¹{price}")
            count += 1

    while True:
        choice = input("\nEnter the item number you want to order (or type 'done' to finish): ").strip().lower()
        if choice == "done":
            break
        if choice.isdigit():
            item_index = int(choice) - 1
            if 0 <= item_index < len(menu_list):
                item_name, item_price = menu_list[item_index]
                try:
                    qty = int(input(f"How many {item_name}s would you like? "))
                    if qty <= 0:
                        print("Quantity must be greater than 0.")
                        continue
                    if item_name in order:
                        order[item_name] += qty
                    else:
                        order[item_name] = qty
                except ValueError:
                    print("Please enter a valid number for quantity.")
            else:
                print("Invalid item number.")
        else:
            print("Please enter a valid number.")

    if not order:
        print("âŒ No items selected. Returning to main menu.")
        continue

    Delivery_option = input("\nWould you like Delivery or Pickup? Choose D or P: ").strip().upper()
    if Delivery_option == "D":
        address = input("Please enter your delivery address: ")
        delivery_charge = 40
        delivery_method = "Delivery"
    else:
        address = "Store Pickup"
        delivery_charge = 0
        delivery_method = "Pickup"

    # --- Receipt Section ---
    print("\n" + "=" * 40)
    print(f"{'GOLDEN FORK RECEIPT':^40}")
    print("=" * 40)
    print(f"Order ID: #{order_id}")
    print(f"Customer Name: {name}")
    print(f"Delivery Method: {delivery_method}")
    print(f"Address: {address}")
    print("-" * 40)
    print(f"{'Item':<15}{'Qty':<5}{'Unit':<7}{'Total':<8}")
    print("-" * 40)

    total = 0
    for item, qty in order.items():
        unit_price = next((menu[cat][item] for cat in menu if item in menu[cat]), 0)
        item_total = unit_price * qty
        total += item_total
        print(f"{item:<15}{qty:<5}{unit_price:<7}â‚¹{item_total:<7}")

    print("-" * 40)
    print(f"{'Delivery Charge':<27}â‚¹{delivery_charge}")
    total += delivery_charge
    print(f"{'Total Amount':<27}â‚¹{total}")
    print("=" * 40)
    print(f"{'Thank you for ordering with us!':^40}")
    print("=" * 40)

    confirm = input("\nDo you want to confirm the order? Choose Y or N: ").strip().upper()
    if confirm == "Y":
        print(f"\nðŸŽ‰ Thank you {name}! Your order #{order_id} has been placed.")
    else:
        print(f"\nâŒ Order #{order_id} cancelled.")

    repeat = input("\nWould you like to place another order? (Y/N): ").strip().upper()
    if repeat != "Y":
        print("ðŸ‘‹ Thank you for visiting Golden Fork! Have a great day!")
        break
```
